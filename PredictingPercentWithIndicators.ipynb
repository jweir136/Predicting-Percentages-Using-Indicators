```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score, learning_curve
from sklearn.preprocessing import StandardScaler

!pip install stockstats

from stockstats import StockDataFrame as Sdf
```

```
FB = pd.read_csv("FB.csv")
FB.head()
```

```
FB_stock = Sdf.retype(FB)
```

```
FB['sma'] = FB_stock['close_10_sma']
FB['boll'] = FB_stock['boll']
```

```
FB['percent'] = ((FB['adj close'] - FB['adj close'].shift(10)) / FB['adj close'].shift(10)) * 100
```

```
FB
```

```
X = FB[['sma', 'boll']].shift(10).dropna()
y = FB['percent'].dropna()

trainX, testX, trainY, testY = train_test_split(X.values, y.values, test_size=.2)
```

```
lr = Pipeline([
    ("scaler", StandardScaler()),
    ("lr", LinearRegression())
])
```

```
score = cross_val_score(lr, trainX, trainY, cv=10, verbose=3, n_jobs=-1, scoring='neg_mean_squared_error')
```

```
np.sqrt(-score.mean()) / FB['adj close'].mean()
```

```
lr.fit(trainX, trainY)

lr.predict([[141.376000, 138.562499	]])
```
